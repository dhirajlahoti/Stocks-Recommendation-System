```python
import datetime as dt
import talib as ta
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
```

```python
import requests
import pandas_datareader as web

session = requests.session()
session.headers['User-Agent'] = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'

data = web.DataReader('AAPL', data_source='yahoo', session=session)
data
```

```python
data['SMA_100'] = ta.SMA(data['Close'],100)
```

```python
data
```

```python
data.iloc[100:]
```

```python
plt.plot(data['Close'])
plt.plot(data['SMA_100'])
```

```python
data['EMA_100'] = ta.EMA(data['Close'], 100)
```

```python
plt.plot(data['Close'])
plt.plot(data['EMA_100'])
```

```python
data['RSI'] = ta.RSI(data['Close'])
```

```python
fig, axs = plt.subplots(2, 1, gridspec_kw={"height_ratios": [3, 1]}, figsize=(10, 6))

axs[0].plot(data['Close'])
axs[1].axhline(y=70, color="r", linestyle="--")
axs[1].axhline(y=30, color="g", linestyle="--")
axs[1].plot(data['RSI'], color="orange")
```

```python
data['LINEARREG'] = ta.LINEARREG(data['Close'], 200)
```

```python
plt.plot(data['Close'])
plt.plot(data['LINEARREG'])
```

```python
start = dt.datetime(2021, 1, 1)
end = dt.datetime.now()

data = web.DataReader('GS', data_source='yahoo', session=session)
macd, macd_signal, macd_hist = ta.MACD(data['Close'])

fig, axs = plt.subplots(2, 1, gridspec_kw={"height_ratios": [5, 2]}, figsize=(25, 12))
c = ["red" if cl < 0 else "green" for cl in macd_hist]
axs[0].plot(data['Close'])
axs[1].plot(macd, 'b-')
axs[1].plot(macd_signal, '--', color='orange')
axs[1].bar(macd_hist.index,macd_hist, color=c)
```
