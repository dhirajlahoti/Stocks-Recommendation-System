```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yf.Ticker('AAPL').info
```

```python
tickers = ['AAPL', 'MSFT', 'IBM']
```

```python
infos = []

for i in tickers:
  infos.append(yf.Ticker(i).info)
```

```python
fundamentals = ['dividendYield', 'marketCap', 'beta', 'forwardPE']
```

```python
df = pd.DataFrame(infos)
df
```

```python
df.columns.isin(fundamentals)
```

```python
df2 = pd.read_html('https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average')[1]
df2
```

```python
tickers = df2.Symbol.to_list()
tickers
```

```python
infos= []

for i in tickers:
  infos.append(yf.Ticker(i).info)
```

```python
df2 = pd.DataFrame(infos)
df2
```

```python
df2 = df2.set_index('symbol')
```

```python
df2=df2.sort_values(by='dividendYield',ascending=False)
df2=df2.drop(columns='companyOfficers')
df2
```

```python
plt.figure(figsize=(14,5))
plt.bar(df2['shortName'],df2['dividendYield'])
plt.xticks(rotation=90)
plt.show
```

```python
df2=df2.sort_values(by='beta',ascending=False)
plt.figure(figsize=(14,5))
plt.bar(df2['shortName'],df2['beta'],color='green')
plt.xticks(rotation=90)
plt.show
```

```python
df2=df2.sort_values(by='marketCap',ascending=False)
plt.figure(figsize=(14,5))
plt.bar(df2['shortName'],df2['marketCap'], color= 'cyan')
plt.xticks(rotation=90)
plt.show
```

```python
df2=df2.sort_values(by='forwardPE',ascending=True)
plt.figure(figsize=(14,5))
plt.bar(df2['shortName'],df2['forwardPE'], color='red')
plt.xticks(rotation=90)
plt.show
```

```python
df2=df2.sort_values(by='forwardEps',ascending=False)
plt.figure(figsize=(14,5))
plt.bar(df2['shortName'],df2['forwardEps'], color='black')
plt.xticks(rotation=90)
plt.show
```

```python
plt.figure(figsize=(14,5))
plt.bar(df2['shortName'],df2['trailingEps'])
plt.xticks(rotation=90)
plt.show
```
